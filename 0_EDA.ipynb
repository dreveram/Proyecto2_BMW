## EDA

#### Primera visualización

```python
# 1. Información general del DataFrame
def general_info(df):
    print("Información General del DataFrame:")
    print(df.info())
    print("\nPrimeras 5 filas:")
    print(df.head())
    print("\nÚltimas 5 filas:")
    print(df.tail())
    print("\nNúmero de filas y columnas:", df.shape)

# 2. Descripción estadística del DataFrame
def statistical_summary(df):
    print("\nResumen estadístico de las columnas numéricas:")
    print(df.describe())
    
    print("\nResumen estadístico de las columnas categóricas:")
    print(df.describe(include='object'))

# 3. Valores nulos y duplicados
def missing_duplicates_info(df):
    print("\nValores nulos por columna:")
    print(df.isnull().sum())
    
    print("\nNúmero de filas duplicadas:", df.duplicated().sum())

# 4. Distribución de variables numéricas
def numerical_distribution(df):
    print("\nDistribución de variables numéricas:")
    df.hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
    plt.tight_layout()
    plt.show()

# 5. Distribución de variables categóricas
def categorical_distribution(df):
    print("\nDistribución de variables categóricas:")
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        plt.figure(figsize=(10,5))
        sns.countplot(data=df, x=col, palette='Set2')
        plt.xticks(rotation=45)
        plt.title(f'Distribución de {col}')
        plt.show()

# 6. Outliers utilizando boxplots (solo para variables numéricas)
def outliers_analysis(df):
    print("\nAnálisis de outliers con Boxplots:")
    numerical_cols = df.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        plt.figure(figsize=(10,5))
        sns.boxplot(data=df, x=col, palette='Set3')
        plt.title(f'Distribución y outliers de {col}')
        plt.show()

# Función principal para ejecutar el EDA
def perform_eda(df):
    general_info(df)
    statistical_summary(df)
    missing_duplicates_info(df)
    numerical_distribution(df)
    categorical_distribution(df)
    outliers_analysis(df)

# Llamar a la función para hacer EDA en tu dataframe
perform_eda(df)
```

```python
# KM
df['km'].hist()
```

```python
# imputaré en los nulos de km la media, pero primero outliers
df[df['km']>400000]
```

```python
df.drop(df[df['km']>1000000].index, inplace=True)
df['km'].hist()
```

```python
# POTENCIA
df['potencia'].hist()
```

```python
# Imputaremos la media pero antes de eso outliers
df[(df['potencia']<50) | (df['potencia']>300)] 
```

```python
# La potencia no puede ser 0, le imputo la media del modelo al que corresponde
df['potencia'] = np.where(df['potencia']==0 , df[df['modelo']=='X1']['potencia'].mean(), df['potencia'])
```

```python
# PRECIO
df['precio'].hist()
```

```python
# Imputaremos la media pero antes de eso outliers
df[(df['precio']>100000)] 
```

```python
df.drop(df[df['precio']>100000].index, inplace=True)
df['precio'].hist()
```

```python
# FECHA REGISTRO
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'])
# FECHA VENTA
df['fecha_venta'] = pd.to_datetime(df['fecha_venta'])

# CREAMOS ANTIGUEDAD
df['antiguedad'] = df['fecha_venta'] - df['fecha_registro'] # está en días, pasamos a años
df['antiguedad'] = ((df['antiguedad'] / np.timedelta64(1, 'D')) / 365.25).round(2)

# ESTUDIO DE ANTIGUEDAD - buscamos valores raros
df['antiguedad'].describe().T
df[df['antiguedad']<=0]
```

```python
# Sustituimos valores outliers de antiguedad por -1
df['antiguedad'] = np.where(df['antiguedad']<0, -1, df['antiguedad'])
```

```python
# ELIMINAMOS COLUMNAS FECHA
del df['fecha_registro']
del df['fecha_venta']
```

```python
df.fillna({
    'modelo':'unknown',
    'km':df['km'].mean(),
    'potencia':df['potencia'].mean(),
    'tipo_gasolina': 'unknown',
    'color': "unknown",
    'tipo_coche': 'unknown',
    'volante_regulable':'unknown',
    'aire_acondicionado':'unknown',
    'camara_trasera':'unknown',
    'asientos_traseros_plegables':'unknown',
    'elevalunas_electrico':'unknown',
    'bluetooth':'unknown',
    'alerta_lim_velocidad':'unknown',
    'precio':df['precio'].mean(),
    'antiguedad':-1     
}, inplace = True)

print(df.isnull().sum())
```
